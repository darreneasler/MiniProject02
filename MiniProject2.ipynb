```python
#1 Load and Inspect the Data (Loading Data)

import numpy as np
import matplotlib as plt
import pandas as pd
import gdown
print("pandas", pd.__version__)
print()
print("numpy", np.__version__)
print()
print("matplotlib", plt.__version__)
print()

csv_id = "1a-9znIQRtfbcEQnPLT957bGmBZL5bNew"
url = f"https://drive.google.com/uc?id={csv_id}"
gdown.download(url, "telco.csv", quiet=False)

print()
dataset = pd.read_csv("telco.csv")
print(dataset.shape)
print()
dataset.head()


```

```python
#1 Load and Inspect the Data (Verify Data)

print(dataset.isnull().sum())
print()
print(f"Target (Monthly Charges) is type: {dataset['target'].dtypes}")

dataset = dataset.rename(columns = {"InternetService_Fiber optic": "InternetService_FO"})
```

```python
#2 Define the Variables (Setting Target and Predictors)

dataset = dataset.rename(columns = {"target": "monthly_charges"})

predictors = ["PhoneService", "MultipleLines", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies", "InternetService_FO", "InternetService_No"]
target = ["monthly_charges"]
delete_columns = dataset.columns.difference(predictors + target)

dataset = dataset.drop(columns = delete_columns)
dataset = dataset[["PhoneService", "MultipleLines", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies", "InternetService_FO", "InternetService_No", "monthly_charges"]]
dataset = dataset.rename(columns = {"PhoneService": "phone_service", "MultipleLines": "multiple_lines", "OnlineSecurity": "online_security", "OnlineBackup": "online_backup", "DeviceProtection": "device_protection", "TechSupport": "tech_support", "StreamingTV": "streaming_tv", "StreamingMovies": "streaming_movies", "InternetService_FO": "internet_fo", "InternetService_No": "internet_no"})
print(dataset.head())
print()

```

```python
#3 Split the Data (Random Split 85/15)
from sklearn.model_selection import train_test_split

predictors = ['phone_service', 'multiple_lines', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies', 'internet_fo', 'internet_no']
target = ['monthly_charges']

X = dataset[predictors]
Y = dataset[target]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
#4 Fit a Multivariable Linear Regression (Y = w_0 + sum(w_i*X_i))

from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(X_train,Y_train)
```

```python
#5 Report the Parameters (Intercept and 10 Linear Coefficients)

intercept = float(model.intercept_)
print(f"Base Rate (Intercept): ${intercept:.2f} / mo")
print()

linear_coefficients = (pd.DataFrame({'service': predictors, '$': model.coef_[0]}).sort_values('$', ascending=False))
linear_coefficients
```

```python
#6 Evaluate on the Test Set (Model Verification)

from sklearn.metrics import r2_score

Y_prediction = model.predict(X_test)

results = X_test.copy()
results = results.drop(columns = predictors)
results['actual'] = Y_test.values
results['model_prediction'] = Y_prediction
results['error'] = abs((results['actual'] - results['model_prediction']))
results['error_squared'] = results['error']**2
print(results.round(2).head(15))
print()

mae = (1/len(Y_test)) * results['error'].sum()
rmse = np.sqrt((1/len(Y_test)) * results['error_squared'].sum())
r_2 = r2_score(Y_test, Y_prediction)
print(f"Mean absolute error: {mae:.2f}")
print()
print(f"Root mean square error: {rmse:.2f}")
print()
print(f"R-Squared: {r_2:.5f}")
```

```python
#7 Compare to a Baseline (Single Variable Linear Regression)

addon_cols = ['phone_service', 'multiple_lines', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies', 'internet_fo', 'internet_no']
df = pd.DataFrame({'num_addons': dataset[addon_cols].sum(axis=1),   'monthly_charge': dataset['monthly_charges']})

X1 = df[["num_addons"]]
Y1 = df[["monthly_charge"]]

X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X1, Y1, test_size = 0.15, random_state = 42)

print("X_train shape:", X_train1.shape)
print("X_test shape:", X_test1.shape)
print("Y_train shape:", Y_train1.shape)
print("Y_test shape:", Y_test1.shape)

model1=linear_model.LinearRegression()
model1.fit(X_train1,Y_train1)

coef = float(model1.coef_[0])
intb = float(model1.intercept_)


print()
print(f"Single Coefficient: {coef:.2f}")
print(f"Single Intercept: {intb:.2f}")
print(f"Linear Regression: {coef:.2f} * num_addons + {intb:.2f}")

Y1_prediction = model1.predict(X_test1)

results1 = X_test1.copy()
results1['actual'] = Y_test1.values
results1['model_prediction'] = Y1_prediction
results1['error'] = abs((results1['actual'] - results1['model_prediction']))
results1['error_squared'] = results1['error']**2
print(results1.round(2).head(15))
print()

mae = (1/len(Y_test1)) * results1['error'].sum()
rmse = np.sqrt((1/len(Y_test1)) * results1['error_squared'].sum())
r_2 = r2_score(Y_test1, Y1_prediction)
print(f"Mean absolute error: {mae:.2f}")
print()
print(f"Root mean square error: {rmse:.2f}")
print()
print(f"R-Squared: {r_2:.5f}")


```
